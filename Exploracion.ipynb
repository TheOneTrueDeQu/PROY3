```python
import pandas as pd

# Especifica la ruta al archivo CSV
file_path = "datos_reemplazados_no_estesi.csv"  # Cambia esto por la ruta real de tu archivo

# Cargar el archivo CSV en un DataFrame
df = pd.read_csv(file_path)

# Mostrar las primeras filas del DataFrame
print(df.head())
```

```python
column_headers = list(df.columns)

# Mostrar los encabezados
print("Encabezados del archivo CSV:", column_headers)
```

```python
data_num = df[["PUNT_INGLES","PUNT_MATEMATICAS","PUNT_SOCIALES_CIUDADANAS", "PUNT_C_NATURALES", "PUNT_LECTURA_CRITICA","PUNT_GLOBAL"]]
data_cat = df[['COLE_BILINGUE', 'COLE_CALENDARIO', 'COLE_CARACTER', 'COLE_JORNADA', 'COLE_NATURALEZA', 'ESTU_GENERO', 'FAMI_EDUCACIONMADRE', 'FAMI_EDUCACIONPADRE', 'FAMI_ESTRATOVIVIENDA', 'FAMI_TIENEAUTOMOVIL', 'FAMI_TIENECOMPUTADOR', 'FAMI_TIENEINTERNET', 'FAMI_TIENELAVADORA']]
data_num.describe()
```

```python
sns.pairplot(data_num)
```

```python
corr = data_num.corr()
sns.heatmap(corr, cmap="Blues", annot=True, fmt=".2f")
```

```python
# Crear boxplots y violin plots para cada columna del DataFrame
for column in data_num.columns:
    plt.figure(figsize=(10, 5))
    
    # Boxplot
    plt.subplot(1, 2, 1)
    sns.boxplot(y=data_num[column])
    plt.title(f'Boxplot of {column}')
    
    # Violin plot
    plt.subplot(1, 2, 2)
    sns.violinplot(y=data_num[column])
    plt.title(f'Violin Plot of {column}')
    
    # Mostrar la gráfica
    plt.tight_layout()
    plt.show()
```

```python
cat_info = ['COLE_BILINGUE', 'COLE_CALENDARIO', 'COLE_CARACTER', 'COLE_JORNADA', 'COLE_NATURALEZA', 'ESTU_GENERO', 'FAMI_EDUCACIONMADRE', 'FAMI_EDUCACIONPADRE', 'FAMI_ESTRATOVIVIENDA', 'FAMI_TIENEAUTOMOVIL', 'FAMI_TIENECOMPUTADOR', 'FAMI_TIENEINTERNET', 'FAMI_TIENELAVADORA']
num_info = ["PUNT_INGLES","PUNT_MATEMATICAS","PUNT_SOCIALES_CIUDADANAS", "PUNT_C_NATURALES", "PUNT_LECTURA_CRITICA"]
global_info = ["PUNT_GLOBAL"]
```

```python
import matplotlib.pyplot as plt

# Crear gráficos de barras agrupadas para cada categoría en cat_info
for category in cat_info:
    # Agrupar por la categoría y calcular la media para cada columna numérica
    grouped_means = df.groupby(category)[num_info].mean()

    # Crear el gráfico de barras agrupadas
    grouped_means.plot(kind='bar', figsize=(12, 6), alpha=0.8, width=0.8)
    
    # Configurar etiquetas y título
    plt.title(f"Promedio de puntajes por {category}", fontsize=14)
    plt.xlabel(category, fontsize=12)
    plt.ylabel("Media de Puntajes", fontsize=12)
    plt.xticks(rotation=45, ha='right')
    plt.legend(title="Variables Numéricas", bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.grid(axis='y', linestyle='--', alpha=0.6)
    
    # Mostrar el gráfico
    plt.tight_layout()
    plt.show()
```
