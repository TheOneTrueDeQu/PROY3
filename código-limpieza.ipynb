```python
filtro = df["COLE_COD_DEPTO_UBICACION"].isin([63, 66])
df_filtrado = df[filtro]
df_filtrado.count()
```

```python
columnas_seleccionadas = [
    'COLE_BILINGUE', 'COLE_CALENDARIO', 'COLE_CARACTER',
    'COLE_JORNADA', 'COLE_NATURALEZA', 'ESTU_GENERO',
    'FAMI_EDUCACIONMADRE', 'FAMI_EDUCACIONPADRE', 'FAMI_ESTRATOVIVIENDA',
    'FAMI_TIENEAUTOMOVIL', 'FAMI_TIENECOMPUTADOR',
    'FAMI_TIENEINTERNET', 'FAMI_TIENELAVADORA', 'PUNT_INGLES',
    'PUNT_MATEMATICAS', 'PUNT_SOCIALES_CIUDADANAS', 'PUNT_C_NATURALES',
    'PUNT_LECTURA_CRITICA', 'PUNT_GLOBAL'
]

# Filtrar el DataFrame conservando solo las columnas seleccionadas
df_filtrado = df_filtrado[columnas_seleccionadas]
```

```python
valores_nulos_por_columna = df_filtrado.isnull().sum()
print("Valores nulos por columna:")
print(valores_nulos_por_columna)
```

```python
df_sin_nulos = df_filtrado.dropna(subset=['PUNT_GLOBAL'])
```

```python
valores_nulos_por_columna = df_sin_nulos.isnull().sum()
print("Valores nulos por columna:")
print(valores_nulos_por_columna)
```

```python
columnas_a_reemplazar = [
    'FAMI_TIENEAUTOMOVIL',
    'FAMI_TIENECOMPUTADOR', 'FAMI_TIENEINTERNET', 'FAMI_TIENELAVADORA'
]

# Reemplazar NaN por "No" en las columnas especificadas
df_sin_nulos[columnas_a_reemplazar] = df_sin_nulos[columnas_a_reemplazar].fillna("No")
```

```python
valores_nulos_por_columna = df_sin_nulos.isnull().sum()
print("Valores nulos por columna:")
print(valores_nulos_por_columna)
```

```python
# Reemplazar NaN por "No sabe" en FAMI_EDUCACIONMADRE y FAMI_EDUCACIONPADRE
df_sin_nulos['FAMI_EDUCACIONMADRE'] = df_sin_nulos['FAMI_EDUCACIONMADRE'].fillna("No sabe")
df_sin_nulos['FAMI_EDUCACIONPADRE'] = df_sin_nulos['FAMI_EDUCACIONPADRE'].fillna("No sabe")

# Reemplazar NaN por "Estrato 2" en FAMI_ESTRATOVIVIENDA
df_sin_nulos['FAMI_ESTRATOVIVIENDA'] = df_sin_nulos['FAMI_ESTRATOVIVIENDA'].fillna("Estrato 2")
df_sin_nulos['COLE_BILINGUE'] = df_sin_nulos['COLE_BILINGUE'].fillna("N")
```

```python
valores_nulos_por_columna = df_sin_nulos.isnull().sum()
print("Valores nulos por columna:")
print(valores_nulos_por_columna)
```

```python
df_sin_nulos = df_sin_nulos.dropna(subset=['PUNT_INGLES'])
```

```python
df_sin_nulos['COLE_CARACTER'] = df_sin_nulos['COLE_CARACTER'].fillna("NO APLICA")
df_sin_nulos['ESTU_GENERO'] = df_sin_nulos['ESTU_GENERO'].fillna("No especifíca")
```

```python
df_sin_nulos = df_sin_nulos.dropna()
```
